```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
driver=webdriver.Chrome('chrome.exe')
driver.get("https://www.youtube.com")
```

```python
# click manually on any video on youtube to scrap comments,upvote,time of comment

comment=[]
timing=[]
upvotes=[]
for _ in range(20):
    driver.execute_script("window.scrollBy(0,1000)")
    time.sleep(2) 
    comments=driver.find_elements(By.XPATH,'//div[@class="style-scope ytd-comment-renderer"]/div[2]/div[2]')
    comment_upvotes=driver.find_elements(By.XPATH,'//div[@class="style-scope ytd-comment-renderer"]//span[@id="vote-count-middle"]')
    for j,upvote in zip(comments,comment_upvotes):
        comment.append(j.text)
        upvotes.append(upvote.text)
        if len(comment)>=500:
            break
    times=driver.find_elements(By.XPATH,'//div[@class="style-scope ytd-comment-renderer"]/div[2]/div/div[2]/yt-formatted-string')
    for tm in times:
        timing.append(tm.text)
        if len(timing)>=500:
            break
    if len(comment)==500 and len(timing)==500:
        break
```

```python
len(timing),len(comment),len(upvotes)
```

```python
df=pd.DataFrame({'youtube video comments':comment,'comment time':timing,'upvotes':upvotes})
df
```

```python
driver=webdriver.Chrome('chrome.exe')
driver.get("https://www.hostelworld.com")
```

```python
search_location=driver.find_element(By.XPATH,'//input[@id="search-input-field"]')
search_location.send_keys(input("enter location name-"))
```

```python
click_on_dropdown=driver.find_element(By.XPATH,'//ul[@id="predicted-search-results"]/li[2]/div')
click_on_dropdown.click()
```

```python
click_on_dropdown=driver.find_element(By.XPATH,'//button[@class="button primary large"]')
click_on_dropdown.click()
```

```python
url=[]
hostel_name=[]
distance_from_citycentre=[]
ratings=[]
totalreviews=[]
overallreview=[]
privates=[]
dorms=[]
facilities=[]
property_description=[]

hostel=driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
for hos in hostel:
    hostel_name.append(hos.text)

distance=driver.find_elements(By.XPATH,'//span[@class="description"]/span')
for dis in distance:
    distance_from_citycentre.append(dis.text)
    
rating=driver.find_elements(By.XPATH,'//div[@class="rating rating-summary-container big"]/div[1]')
for rat in rating:
    ratings.append(rat.text) 
    
totalreview=driver.find_elements(By.XPATH,'//div[@class="reviews"]')
for rev in totalreview:
    totalreviews.append(rev.text)
    
overallreviews=driver.find_elements(By.XPATH,'//div[@class="keyword"]')
for ovr in overallreviews:
    overallreview.append(ovr.text) 
    
private=driver.find_elements(By.XPATH,'//div[@class="price-col"][1]')
for pri in private:
    privates.append(pri.text)    
    
dorm=driver.find_elements(By.XPATH,'//div[@class="price-col"][2]')
for dor in dorm:
    dorms.append(dor.text) 
    
facility=driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
for fac in facility:
    facilities.append(fac.text)

link=driver.find_elements(By.XPATH,'//h2[@class="title title-6"]/a')
for i in link:
    url.append(i.get_attribute('href'))
for j in url:
    driver.get(j)
    propert2=driver.find_elements(By.XPATH,'//div[@class="flex-80"]/div/div/div[2]')
    for prop in propert2:
        property_description.append(prop.text)    
        
        
```

```python
print(len(hostel_name),len(distance_from_citycentre),len(ratings),len(totalreviews),len(overallreview),len(privates),len(dorms),len(facilities),len(property_description))
```

```python
df=pd.DataFrame({'Hostel':hostel_name,'Distance_from_citycentre':distance_from_citycentre,'Ratings':ratings,'totalreviews':totalreviews,'overallreview':overallreview,'privates fees':privates,'dorms fees':dorms,'facilities':facilities,'property_description':property_description})
df
```
