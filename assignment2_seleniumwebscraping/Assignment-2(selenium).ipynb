```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
driver.get("https://www.flipkart.com/")
```

```python
search=driver.find_element(By.CLASS_NAME,"_3704LK")
search.send_keys('sunglasses')
```

```python
search_click=driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search_click.click()
```

```python
brand=[]
productdescription=[]
price=[]
offers=[]
```

```python
start=0
end=3
for page in range(start,end):

    brand_tag=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')         # scraping brand
    for i in brand_tag:
        brands=i.text
        brand.append(brands)
    
    productdescription_tag=driver.find_elements(By.CLASS_NAME,'IRpwTa')   # scraping productdescription
    for i in productdescription_tag:
        productdescriptions=i.text
        productdescription.append(productdescriptions) 

    price_tag=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')           # scraping price
    for i in price_tag:
        prices=i.text
        price.append(prices) 

    offers_tag=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')        # scraping offers
    for i in offers_tag:
        offerss=i.text
        offers.append(offerss)
    next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
    next_button.click()
    time.sleep(1)
```

```python
print(len(brand),len(productdescription),len(price),len(offers))
```

```python
df=pd.DataFrame({'Brand':brand[0:100],'productdescription':productdescription[0:100],'price':price[0:100],'offers':offers[0:100]})
df
```

```python

quotes=[]
author=[]
typeofquote=[]
for page in range(0,4):
    quotes_tag=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quotes_tag:
        quote=i.text
        quotes.append(quote)

    author_tag=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author_tag:
        authors=i.text
        author.append(authors)

    typeofquote_tag=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in typeofquote_tag:
        typeofquotes=i.text
        typeofquote.append(typeofquotes)
        
    next_button=driver.find_element(By.XPATH,'//li[@class="next"]')
    next_button.click()
    
```

```python

sn_no=[]
name=[]
born_dead=[]
term=[]
remark=[]

sn_no_tag=driver.find_elements(By.XPATH,'//div[@class="table-box"][1]//tr')
for i in sn_no_tag:
    sn_nos=i.text.split('\n')[0]
    sn_no.append(sn_nos)

name_tag=driver.find_elements(By.XPATH,'//div[@class="table-box"][1]//tr')
for i in name_tag:
    names=i.text.split('\n')[1]
    name.append(names)

born_dead_tag=driver.find_elements(By.XPATH,'//div[@class="table-box"][1]//tr')
for i in born_dead_tag:
    born_deads=i.text.split('\n')[2]
    born_dead.append(born_deads)
    
term_tag=driver.find_elements(By.XPATH,'//div[@class="table-box"][1]//tr')
for i in term_tag:
    terms=i.text.split('\n')[3]
    term.append(terms)
    
remark_tag=driver.find_elements(By.XPATH,'//div[@class="table-box"][1]//tr')
for i in remark_tag:
    remarks=i.text.split('\n')[4]
    remark.append(remarks)    

    
    
```

```python
print(len(sn_no),len(name),len(born_dead),len(term),len(remark))
```

```python
df=pd.DataFrame({"SN.no":sn_no[1:],"NAME":name[1:],"BORN-DEAD":born_dead[1:],"TERM":term[1:],"REMARK":remark[1:]})
df
```
