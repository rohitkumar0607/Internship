```python
from bs4 import BeautifulSoup
import requests
page=requests.get("https://www.imdb.com/chart/top/?sort=ir,desc&mode=simple&page=1")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
imdb_top_50_movies=[]
for i in soup.find_all('td',class_="titleColumn"):
    imdb_top_50_movies.append(i.text.split('\n')[2])
    
imdb_top_50_movies
```

```python
imdb_top_50_movies_rating=[]
for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    imdb_top_50_movies_rating.append(i.text.split('\n')[1])
    
imdb_top_50_movies_rating
```

```python
imdb_top_50_movies_yearofrelease=[]
for i in soup.find_all('span',class_="secondaryInfo"):
    imdb_top_50_movies_yearofrelease.append(i.text)
    
imdb_top_50_movies_yearofrelease
```

```python
print(len(imdb_top_50_movies),len(imdb_top_50_movies_rating),len(imdb_top_50_movies_yearofrelease))
```

```python
import pandas as pd
df=pd.DataFrame({"IMDB TOP 50 MOVIES":imdb_top_50_movies[0:50],"RATINGS":imdb_top_50_movies_rating[0:50],"YEAR":imdb_top_50_movies_yearofrelease[0:50]})
df
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
top10_odi_team=[]
for i in soup.find_all('span', class_="u-hide-phablet"):
    top10_odi_team.append(i.text)
    
top10_odi_team 
```

```python
top10_odi_team_matches=[]
for i in soup.find_all('td', class_="rankings-block__banner--matches"):
    top10_odi_team_matches.append(i.text)
top10_odi_team_match=[]
for i in soup.find_all('td', class_="table-body__cell u-center-text"):
    top10_odi_team_match.append(i.text.split(',')[0])
top10_odi_team_match=top10_odi_team_match[0::2]
top10_odi_team_matches.extend(top10_odi_team_match)
top10_odi_team_matches
```

```python
top10_odi_team_points=[]
for i in soup.find_all('td', class_="rankings-block__banner--points"):
    top10_odi_team_points.append(i.text)
top10_odi_team_point1=[]
for i in soup.find_all('td', class_="table-body__cell u-center-text"):
    top10_odi_team_point1.append(i.text)
top10_odi_team_point1=top10_odi_team_point1[1::2]
top10_odi_team_points.extend(top10_odi_team_point1)
top10_odi_team_points
```

```python
top10_odi_team_ratings=[]
for i in soup.find_all('td', class_="rankings-block__banner--rating u-text-right"):
    top10_odi_team_ratings.append(i.text.split('\n')[1])
top10_odi_team_ratings1=[]
for i in soup.find_all('td', class_="table-body__cell u-text-right rating"):
    top10_odi_team_ratings1.append(i.text)
top10_odi_team_ratings.extend(top10_odi_team_ratings1)
top10_odi_team_ratings
```

```python
print(len(top10_odi_team),len(top10_odi_team_matches),len(top10_odi_team_points),len(top10_odi_team_ratings))
```

```python
import pandas as pd
df=pd.DataFrame({"Top10 mens odi team":top10_odi_team[:10],"Matches":top10_odi_team_matches[:10],"points":top10_odi_team_points[:10],"Ratings":top10_odi_team_ratings[:10]})
df
```

```python
Top10_odi_batsmen_rating=[]
for i in soup.find_all("div",class_="rankings-block__banner--rating"):
    Top10_odi_batsmen_rating.append(i.text)
top10_odi_batsmen_rating1=[]
for i in soup.find_all('td', class_="table-body__cell rating"):
    top10_odi_batsmen_rating1.append(i.text)
Top10_odi_batsmen_rating.extend(top10_odi_batsmen_rating1)
Top10_odi_batsmen_rating
```

```python
print(len(Top10_odi_batsmen),len(Top10_odi_batsmen_team),len(Top10_odi_batsmen_rating))
```
